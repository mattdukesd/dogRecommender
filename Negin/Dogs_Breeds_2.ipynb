```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
my_str=""
```

```python
html = requests.get('http://petadoption.ocpetinfo.com/LostFound/FoundList.asp?type=Dog')
# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(html.text, 'html.parser')
# Examine the results, then determine element that contains sought info
# results are returned as an iterable list      
                   
table=soup.find_all('table',class_="tblpetbrdr")    
```

```python
for tr in soup.find_all('tr'):
    link=tr.find('a', {'href': re.compile('LFShowImage.asp?')})   
```

```python
str_list=""

for tab in table:
       str_list+=tab.text
```

```python
images = soup.findAll('img')
img_list=[]
link_list=[]
for image in images:
    img_list.append(image['src'])
```

```python
unwanted = {"/ocgov/img/blank.gif","images/camera.gif","images/back-to-search.gif"}
img_list = [e for e in img_list if e not in unwanted]
```

```python
img_list

for link in img_list:
    
    link_list.append(link.replace("../","http://petadoption.ocpetinfo.com/"))
```

```python
link_list
```

```python
newStr =str_list .split('\n')
newstr=[x for x in newStr if x!='\r' ]
newstr=[x for x in newstr if x!='\n']
newstr=[x for x in newstr if x!='']
newstr=[s.replace('                  ', '') for s in newstr] 
newstr = [s.replace('\r', '') for s in newstr]
newstr=[s.replace('\xa0', ' ') for s in newstr] 
newstr=[s.replace('\xa0', ' ') for s in newstr] 

newstr=[s.replace('              ', '') for s in newstr]    
```

```python
for j in range(len(newstr)-1):
    if (newstr[j]==' ' and newstr[j+1]!=''):
         newstr[j]="*"
newstr=[x for x in newstr if x!='*']
newstr=[x for x in newstr if x!=' ']
newstr=[x for x in newstr if x!='            ']
newstr=[x for x in newstr if x!='\t\t\t\t  ']
```

```python
composite_list = [newstr[x:x+8] for x in range(0, len(newstr),8)]
```

```python
dt=pd.DataFrame(composite_list, columns = composite_list[0])
dt
```

```python

dt=dt.drop([0])
dt['link']=link_list
dt.info()
```
